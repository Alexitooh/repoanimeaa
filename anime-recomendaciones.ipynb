```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


hunter_id = 136
school_days_id = 2476   
# CARGAR LOS DATOS
try:
    anime_df = pd.read_csv('../data/anime.csv')
    ratings_df = pd.read_csv('../data/rating.csv')
except FileNotFoundError:
    print("‚ùå Error: No encuentro los archivos")
    print("Aseg√∫rate de que 'anime.csv' y 'rating.csv' est√©n en la carpeta 'datos'")

# EXPLORACION DE DATOS ANTES
# -----------------------------ANIME---------------------------------------------- 
print("\n--- EXPLORANDO DATOS DE ANIME ---")
print(f"\nüìä Informaci√≥n general del dataset de anime:")
print(f"‚Ä¢ N√∫mero de animes: {len(anime_df)}")
print(f"‚Ä¢ Columnas de la TABLA ANIMES : {list(anime_df.columns)}")
print(f"‚Ä¢ Tipos de datos de la Tabla:")
print(anime_df.dtypes)

#FALTAN CONTAR LOS ANIMES DESPUES DE LOS FILTROS y ANTES 

# # ----------------------RATINGS---------------------------------------------------- 
print("--- EXPLORANDO DATOS DE RATINGS ---")
print(f"\nüìä Informaci√≥n de ratings:")
print(f"‚Ä¢ Total de calificaciones: {len(ratings_df):,}")
print(f"‚Ä¢ Usuarios: {ratings_df['user_id'].nunique():,}")
print(f"‚Ä¢ Columnas de la TABLA RATINGS : {list(ratings_df.columns)}")
print(ratings_df.dtypes)

#-----------------LIMPIAR DATOS------------------------------------------------------

#ELIMINAR ANIMES NO CALIFICADOS
# Aqui me interesa quitar a las personas que no han visto la serie  ya que no tienen uso en el sistema de recomendaciones 
ratings_clean = ratings_df[ratings_df['rating'] != -1]
#Cuantas Calificaciones quedan despues del Filtro
print(f"Calificaciones despu√©s de eliminar -1: {len(ratings_clean):,}")

# ELIMINAR ANIME CON POCAS CALIFICACIONES  Y USUARIOS QUE HAYAN CALIFICADO POCO
min_anime_ratings = 50
min_user_ratings = 10
anime_counts = ratings_clean['anime_id'].value_counts()
user_counts = ratings_clean['user_id'].value_counts()

popular_animes = anime_counts[anime_counts >= 50].index
active_users = user_counts[user_counts >= 10].index

#POPULARES
print(len(popular_animes))
print(len(active_users))

ratings_filtered = ratings_clean[
    ratings_clean['anime_id'].isin(popular_animes) &  # "Mant√©n solo las filas donde el anime_id est√© en la lista de populares"
    ratings_clean['user_id'].isin(active_users)       # Mant√©n solo las filas donde el user_id est√© en la lista de activos
]


print(f"   ‚Ä¢ Ratings DESPU√âS: {len(ratings_filtered):,}")


```
